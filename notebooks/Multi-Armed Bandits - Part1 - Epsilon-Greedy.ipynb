```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt


from samples.rl.bandits_epsilon_greedy import (
#     Bandit,
#     EpsilonGreedyAgent,
    compare_epsilons,
)

%matplotlib inline
```

```python
epsilons = [0.01, 0.05, 0.1, None]
bandits_means = [3, 4, 5]

iterations = 50
agents, _ =  compare_epsilons(epsilons, bandits_means, iterations)

all_rewards = [agent.rewards_log.all_rewards for agent in agents]

for i in range(len(agents)):
    plt.plot(
        np.cumsum(all_rewards[i]) / np.arange(1, iterations + 1), 
        label="epsilon = {}".format(epsilons[i] or '1 / iter')
    )

plt.xlabel("iteration")
plt.ylabel("average cumulative rewards")
plt.legend()
plt.show()
```

```python
for i in range(len(agents)):
    plt.plot(
        np.cumsum(all_rewards[i]) / np.arange(1, iterations + 1), 
        label="epsilon = {}".format(epsilons[i] or '1 / iter'),
    )

plt.xlabel("log-iteration")
plt.ylabel("average rewards")
plt.xscale("log")
plt.legend()
plt.show()
```

```python
for i in range(len(agents)):
    plt.plot(
        np.cumsum(all_rewards[i]), 
        label="epsilon = {}".format(epsilons[i] or '1 / iter'),
    )

plt.xlabel("iteration")
plt.ylabel("total rewards")
plt.legend()
plt.show()
```
